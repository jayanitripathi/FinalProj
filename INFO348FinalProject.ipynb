## Introduction  

Innovation plays a critical role in shaping economic progress and technological advancements, and understanding the factors that drive it can offer valuable insights. This project aims to analyze the relationship between institutional characteristics, student demographics, and innovation outputs such as patents and citations. Using a dataset that includes information on student income distributions, inventor shares, total patents, and patent citations across various institutions, we explored the correlation between wealth, institutional size, and innovation metrics. Our analysis revealed significant patterns: institutions with a higher proportion of students from wealthier families tend to produce more patents and highly cited innovations. Additionally, a strong positive correlation was observed between total patents and citations, highlighting the consistency in innovation quality. On the contrary, institutions with higher shares of students from lower-income families exhibited lower innovation outcomes. By leveraging data wrangling, visualization, and statistical summarization techniques, this project sheds light on the socioeconomic factors influencing innovation within educational institutions, offering a data-driven perspective on the dynamics of economic opportunity and technological advancement.

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv("innovationrates.csv")
```

##Data Wrangling
1. Remove missing values replaced with NaN.
2. Added a colum which calculates the percentage of students per income quintile.

```python
print("Dataset Overview:")
display(df.head())

print("\nMissing Values:")
print(df.isnull().sum())

cols_to_check = ['inventor', 'inventor_pq1', 'inventor_pq2', 'inventor_pq3', 'inventor_pq4', 'inventor_pq5']
df[cols_to_check] = df[cols_to_check].replace(0, np.nan)

for i in range(1, 6):
    df[f'percent_pq{i}'] = (df[f'count_pq{i}'] / df['count']) * 100

print("\nUpdated Dataset:")
display(df.head())
```

Statitiscal Summarization:

1. Outputs the summary statistics for numerical columns.
2. Find the Average inventor share.
3. Find insitutions with the highest invention rates.
4. Correlation Matrix to find the relationship between variables and see how each variable affects the other.

```python
print("\nSummary Statistics:")
print(df.describe())

avg_inventor_share = df['inventor'].mean()
print(f"\nAverage Share of Inventors: {avg_inventor_share:.4f}")

avg_patents = df['total_patents'].mean()
avg_citations = df['total_cites'].mean()
print(f"Average Total Patents: {avg_patents:.2f}")
print(f"Average Total Citations: {avg_citations:.2f}")

top_inventors = df.sort_values(by='inventor', ascending=False).head(5)
print("\nTop 5 Institutions by Inventor Share:")
display(top_inventors[['instnm', 'inventor']])

print("\nCorrelation Matrix:")

numerical_features = df.select_dtypes(include=np.number)

corr = numerical_features.corr()
display(corr)

plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

#Data Visualization

##Share of Inventors by Institution

California Insitute of Technology has the largest share of inventors.

```python
# Bar chart: Share of inventors by institution
plt.figure(figsize=(12, 6))
sns.barplot(x='inventor', y='instnm', data=df.sort_values('inventor', ascending=False).head(10), palette='viridis')
plt.title('Top 10 Institutions by Invention Rate')
plt.xlabel('Share of Inventors')
plt.ylabel('Institution')
plt.show()
```

##Total Patents vs Total Citations

There is a positive relationship between Total Patents and Total Citations, as the number of patents increase so do the number of citations.

Institutions with high patent counts and citation numbers (top-right of the chart) also tend to have a relatively higher inventor share, suggesting these institutions may provide an environment that supports or encourages innovation.

```python
# Scatter plot: Total Patents vs Total Citations
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_patents', y='total_cites', data=df, hue='inventor', size='inventor', palette='coolwarm', sizes=(20, 200))
plt.title('Total Patents vs Total Citations')
plt.xlabel('Total Patents')
plt.ylabel('Total Citations')
plt.legend(title='Inventor Share')
plt.show()
```

#Inventor Share across Income Quintiles

The share of inventors decreases from Q1 (low income) to Q3 (middle income).
Starting from Q4, the share begins to rise, with Q5 (highest income group) showing the largest proportion of inventors.

The share of inventors is highest among individuals in the top income quintile (Q5), suggesting a positive association between income and participation in inventive activity. Conversely, the middle and lower quintiles show relatively smaller shares of inventors.


```python
# Line plot: Inventor share by income quintile
quintile_cols_inventor = ['inventor_pq1', 'inventor_pq2', 'inventor_pq3', 'inventor_pq4', 'inventor_pq5']
df_mean_inventor = df[quintile_cols_inventor].mean()

plt.figure(figsize=(10, 6))
sns.lineplot(x=quintile_cols_inventor, y=df_mean_inventor.values, marker='o', color='b')
plt.title('Average Inventor Share Across Income Quintiles')
plt.ylabel('Share of Inventors')
plt.xlabel('Income Quintiles')
plt.xticks(range(5), ['Q1', 'Q2', 'Q3', 'Q4', 'Q5'])
plt.show()
```
